```python
import pandas as pd

# Read the CSV again, telling pandas not to treat the first row as header
df = pd.read_csv(r'E:\jupyter notebook\Capstone\Output(70).csv', header=None)

# Now set the first row as column headers
df.columns = df.iloc[0]  # First row becomes header
df = df[1:]  # Remove the first row from the data
df.reset_index(drop=True, inplace=True)

df.head()

```

```python
import pandas as pd

# Read the CSV using default header (first row as column names)
df = pd.read_csv(r'E:\jupyter notebook\Capstone\Output(70).csv')

df.head()
```

```python
print(len(df.columns.tolist()))
```

```python
# Save original column names
column_names = df.columns.tolist()

# Create a new DataFrame with the column names as the first row
df_with_headers = pd.DataFrame([column_names], columns=column_names)

# Append the original data below it
df = pd.concat([df_with_headers, df], ignore_index=True)

df.head()
```

```python
# Change column names to integers from 1 to 1872
df.columns = list(range(1, df.shape[1] + 1))

df.head()
```

```python
# Rename the last two columns
df.columns = list(df.columns[:-2]) + ['no of rows', 'fall/notfall']

df.head()
```

```python
value = df.loc[0, 'fall/notfall']
print(value)
```

```python
# Change the value in row 0, column 'fall/notfall' to 1
df.loc[0, 'fall/notfall'] = 1

# Save the updated DataFrame to a new CSV file
df.to_csv(r'E:\jupyter notebook\Capstone\Output_1.csv', index=False)
```
