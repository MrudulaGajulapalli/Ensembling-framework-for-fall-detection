```python
import os

base_path = r"E:\Fall Detection Dataset\nonfall_cam1\nfDatasets"

folder_image_counts = {}

# Iterate through each subfolder
for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    
    if os.path.isdir(folder_path):
        # Count number of image files (jpg, png, etc.)
        image_count = sum(1 for file in os.listdir(folder_path) if file.lower().endswith(('.jpg', '.jpeg', '.png')))
        folder_image_counts[folder] = image_count

# Find the folder with the minimum number of images
min_folder = min(folder_image_counts, key=folder_image_counts.get)
min_count = folder_image_counts[min_folder]

# Print results

for folder, count in folder_image_counts.items():
    print(f"Folder: {folder}, Images: {count}")

print(f"\nFolder with minimum images: {min_folder}, Count: {min_count}")
```
